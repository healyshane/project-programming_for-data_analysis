# Background

## Glacier Detail
Most of the world's glacial ice is found in Antarctica and Greenland, but glaciers are found on every continent.

Glaciers store approximately 75 percent of the world's fresh water and 10% of land area on Earth is covered with glacial ice <sup>[1](https://nsidc.org/cryosphere/glaciers/)</sup>.

Image below is of Perito Moreno Glacier in Argentina <sup>[2](https://upload.wikimedia.org/wikipedia/commons/thumb/5/5c/Perito_Moreno_Glacier_Patagonia_Argentina_Luca_Galuzzi_2005.JPG/220px-Perito_Moreno_Glacier_Patagonia_Argentina_Luca_Galuzzi_2005.JPG)</sup>.
![Perito Moreno Glacier](https://upload.wikimedia.org/wikipedia/commons/thumb/5/5c/Perito_Moreno_Glacier_Patagonia_Argentina_Luca_Galuzzi_2005.JPG/1280px-Perito_Moreno_Glacier_Patagonia_Argentina_Luca_Galuzzi_2005.JPG)


Glaciers are dynamic. Snow falls at the glaciers highest elevation. Over time the snow is compacted and turns to ice. As the weight of the ice increases, as a result of gravity, the glacier will slowly flow downhill. 

At a lower altitude down the glacier is the ablation area where majority or melting and evaporation occurs. 

A glacier is in equilibrium when snowfall equals snowmelt. When the equilibrium is disturbed by a change in either snowfall or melting, the glacier either retreats or advances <sup>[1](https://nsidc.org/cryosphere/glaciers/)</sup>. 

There are many different types of glaciers, depending on terrain and scale <sup>[1](https://nsidc.org/cryosphere/glaciers/)</sup> :
* Mountain glaciers
* Valley glaciers
* Tidewater glaciers
* Piedmont glaciers
* Hanging glaciers
* Cirque glaciers
* Ice Aprons
* Rock Glaciers
* Ice shelves
* Icefields
* Ice caps
* Ice streams
* Ice sheets


## Glaciers as a System
Below figure shows the inputs and outs of a glacier system <sup>[3](http://cdn.antarcticglaciers.org/wp-content/uploads/2018/11/glaciers-as-a-system.png)</sup>
![Antarctic Glaciers](http://cdn.antarcticglaciers.org/wp-content/uploads/2018/11/glaciers-as-a-system.png)


For the majority of the world's glaciers, they receive snowfall throughout the winter, and the main ablation occurs in the summer. The Mass Balance, the balance of accumulation and ablation, is usually therefore positive in the winter and negative in the summer <sup>[4](http://www.antarcticglaciers.org/glacier-processes/glacier-accumulation-and-ablation/)</sup>. 

## Why are Glaciers Studied
 

Glaciers are important indicators of global warming and climate change. As Earth’s average temperature has been increasing for more than a century the study and modelling of glaciers is important to highlight risks to Earth's population.  

Melting ice causes rising sea levels. The sea level would rise approximately 70 meters if all the ice covering Antarctica, Greenland, and mountain glaciers around the world was to melt. All coastal cities would be submerged <sup>[5](https://www.amnh.org/explore/ology/earth/ask-a-scientist-about-our-environment/will-the-world-ever-be-all-under-water)</sup>.

Large additions of fresh water from melting glaciers change the ocean ecosystem. Ocean organisms, such as many types of corals, depend on salt water for survival and may not survive a more diluted environment. 

The loss of glacial ice also reduces the amount of fresh water available for humans, animals and plants.

Large icebergs, breaking off glaciers into the sea, create hazards for the shipping industry <sup>[6](https://www.nationalgeographic.org/encyclopedia/glacier/)</sup>.

## Glacier Thickness
In a paper from The Journal of Geophysical Research <sup>[9](https://glaciers.gi.alaska.edu/sites/default/files/wiki/Huss,%20Farinotti%20-%202012%20-%20Distributed%20ice%20thickness%20and%20volume%20of%20all%20glaciers%20around%20the%20globe.pdf)</sup> a method is presented by Matthias Huss and Daniel Farinotti to calculate glacier mean ice thickness. By inverting glacier topography of all glaciers worldwide, details of glacier ice thickness were generated. Results were validated against a set of thickness observations for 300 glaciers around the world.

Ice thickness is measured in meters.

Table 2, Calculated Total Ice Volume and Validation Data for 19 Regions, was copied from the paper referenced above and *Ice Thickness for Regions.csv* file generated. This data file is stored in DATA folder of repository. 


```python
# Import required Python packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math as math
```

```python
# Read in CSV file, Ice Thickness for Regions, from DATA folder in repository
# Create a DataFrame called thick
thick = pd.read_csv("DATA/Ice Thickness for Regions.csv")
thick
```

The root mean square error is around 30%, detailed in the research paper <sup>[9](https://glaciers.gi.alaska.edu/sites/default/files/wiki/Huss,%20Farinotti%20-%202012%20-%20Distributed%20ice%20thickness%20and%20volume%20of%20all%20glaciers%20around%20the%20globe.pdf)</sup>, and includes uncertainties in:
* The thickness modelling
* The glacier outlines and the DEM, digital elevation models, used
* The field data (ground penetrating radar resolution, extrapolation, inconsistency of measurement date and inventory date)


```python
# Column added to DataFrame to provide a standard deviation for height results
thick['Standard Deviation'] = thick['Ice Thickness']*0.3
thick.head(10) # Limit the output to 10 lines
```

```python
# Generate 100 samples of data and hold in new DataFrame
thickness = thick.sample(n=100, replace = True)
thickness.head(10) # Limit the output to 10 lines
```

```python
# Generate a new column in DataFrame to provide 100 data points for ice thickness.
# A normal distribution from the NumPy.random package is used to generate ice thickness data around the mean and standard deviation of original data.
thickness['Ice Thickness_gen'] = np.random.normal(thickness['Ice Thickness'],thickness['Standard Deviation'])
thickness.head(10) # Limit the output to 10 lines 
```

```python
# Create a Pandas DataFrame, data, to hold the data generated during this assignment.
# Round the values
data = thickness[['Ice Thickness_gen']].round()
data.head(10) # Limit the output to 10 lines
```

## Bedrock Conditions
Glaciers flow downhill through the processes of deformation and sliding. One factor that impacts flow is bedrock conditions. The surface over which a glacier is flowing may be hard, soft, frozen or thawed bed <sup>[10](http://www.antarcticglaciers.org/glacier-processes/)</sup>. Given the diversity of regions for which glacier data is being generated, an equal probability of choosing one of the bedrock conditions will be used. 

```python
# Create an array to hold the different bedrock conditions of glaciers.
bedrock = ["hard", "soft", "frozen", "thawed_bed"]

# Generate 100 samples of data from the bedrock arry with equal probability. 
bedrock_data = np.random.choice(bedrock, size=100, replace=True)
bedrock_data
```

```python
# Add the bedrock conditions to the data DataFrame 
data['Bedrock Conditions'] = bedrock_data
data.head(10) # Limit the output to 10 lines
```

## Glacier Slope

The two major factors which control the advance/retreat of the Himalayan glaciers are  the slope of the glacier and changes in the equilibrium line altitude <sup>[11](https://www.the-cryosphere.net/6/301/2012/tc-6-301-2012.pdf)</sup>. The slope of a glacier is a major factor in influencing how it behaves. 

The Franz Josef Glacier in New Zealand is extremely steep, it descends from 3000m to 240m over 11KM. This makes it one of the steepest glaciers in both New Zealand and the world <sup>[12](https://www.franzjosefglacier.com/franz-josef-glacier/)</sup>.

```python
# Calculate the slope of Franz Josef Glacier as reference point.
max_slope = (3000-240)/11000
print("The mean slope of the Franz Josef Glacier is %3.2f"% max_slope,"= %3.2f" % (90*max_slope),"degrees")
```

In a paper on Minimal Glacier Models <sup>[13](http://www.staff.science.uu.nl/~oerle102/MM2011-all.pdf)</sup>, when assessing the relationship between glacier and air temperature as a function of the bed slope, the minimum glacier slope is defined as 0.05.

As detailed in a paper on Hillslope‐glacier coupling <sup>[14](https://agupubs.onlinelibrary.wiley.com/doi/full/10.1029/2010JF001751)</sup>, High Asian glacial landscapes have large variations in topographic relief and the size and steepness of snow accumulation areas. The mean and standard deviation of glaciers in different regions are referenced. The standard deviation of glacier slopes in six regions were found to be (3.2, 3.4, 2.9, 3.5, 4.1, 4.3) degrees.

```python
# Generate an array to hold the standard deviation values for 6 High Asian glacier regions and calculate the mean.
# The units of result are degrees.
SD_data = np.mean([3.2, 3.4, 2.9, 3.5, 4.1, 4.3])

# Inner brackers - convert degrees to radians
# Outer brackets - results of tan(radians) is decimal percentage 
SD_slope = math.tan(math.radians(SD_data))
SD_slope
```

```python
# Generate an array of size 100 normally distibuted with mean 0.15, standard deviation as calculated by SD_slope.
# Mean of 0.15 was calculated from max and min values of glacier slope as discussed above. 
slope_tmp = np.random.normal(0.15,SD_slope,100)

# As the slope is defined as a decimal of a percentage, multiply by 90 to convert to degrees. 

slope = (slope_tmp*90)
slope

```

```python
# Add the slope to the data DataFrame
data['Glacier Slope'] = slope
data.head(10) # Limit the output to 10 lines
```

## Glen's Flow Law

Glen's flow law describes the relationship between strain and stress of glacial systems.
Glaciers flow as a result of strain in response to stress. Strain may include deformation of the ice or the bedrock or sliding of the glacier. The relation between strain rate and stress is described by Glen’s Flow Law <sup>[15](http://www.antarcticglaciers.org/glacier-processes/glacier-flow-2/glacier-flow-ii-stress-and-strain/)</sup>:


ε = B $\sigma ^ n$ 

Where:

ε = Strain Rate

B = Temperature dependent ice hardness parameter

$\sigma$ = Basal Shear Stress

n = Creep Exponent

### Ice Hardness Parameter

To calculate temperature dependent ice hardness parameter, B, data is specified <sup>[16](https://www.geo.umass.edu/courses/geo563/Geo563-6.IceFlow&profiles.pdf)</sup> at temperatures of (0,-5,-10,-20) degrees celsius as (0.165, 0.054, 0.017, 0.0016) respectively. 

To generate samples from this distribution, a curve will be fitted to the data and an equation to describe the relationship, between temperature and hardness parameter, will be generated. 

```python
# Import SciPy package for curve fitting
from scipy.optimize import curve_fit

# Arrays created to hold data points from literature. 
x= np.array([0, -5,-10,-20]) # temperature data
y1 = np.array([0.165,0.054,0.017,0.0016]) # ice hardness parameter data

# Fit curve to data
p,l = curve_fit(lambda t,a,b: a*np.exp(b*t),  x,  y1)
p,l # Output is two arrays, the first array holds the variables that describe constants in exponential equation.
```

```python
# Define arrays for plotting
x= np.array([0, -5,-10,-20]) # temperature data
y1 = np.array([0.165,0.054,0.017,0.0016]) # actual data
y2 = p[0]*np.exp(p[1]*x) # fitted data from SciPy curve_fit, equation inputs are sliced from generate array

# Generate plot to visualise
plt.scatter(x,y1, label = "Data", color = 'r')
plt.plot(x,y2, label = "Fit", color = 'g')
plt.title("Plot of Data against Exponential Fit")
legend = plt.legend(loc='upper center', shadow=True, fontsize='x-large')
plt.show()
```

The model to describe ice hardness parameter will be used to generate 100 data points with the range of 0 to 20 degrees.

```python
# Generate 100 data points from exponential model between -20degC and 0degC. 
B = p[0]*np.exp(p[1]*np.random.randint(-20,0,100))

plt.hist(B);
```

## Basal Shear Stress

At the base of a glacier with minimal slope, the normal stress acting on the bed is mainly a result of the weight of a glacier. Stress increases with ice thickness and reaches a maximum where the ice meets the bedrock. In simple glaciers, basal shear stress is also controlled by slope. Basal shear stress can be represented by below equation <sup>[15](http://www.antarcticglaciers.org/glacier-processes/glacier-flow-2/glacier-flow-ii-stress-and-strain/)</sup> :

 σ = ρ*g*h*sin(a)

 where:

 σ = Basal Shear Stress
 
 ρ = Ice density
 
 g = gravitation acceleration
 
 h = Ice thickness
 
 a = Glacier slope

```python
# Define variables for use in calculation of basal shear stress
ρ = 900 # density of ice taken as a constant
g = 9.81 # gravity as a constant
h = data['Ice Thickness_gen'] # Generated data for glacier ice thickness
a = np.sin(data['Glacier Slope']) # Generated data for glacier slope

σ = ρ*g*h*a

# Add basal shear stress to the data DataFrame
data['Basal Shear Stress'] = σ
```

```python
# Plot a histogram of generated basal shear stress results to visualise. 
σ.plot(kind='hist');
plt.title("Generate Values of Basal Shear Stress")
plt.xticks(rotation=90)

# Print mean basal shear stress to screen.
print("The mean basal shear stress of data set is %3.0f" %np.mean(σ), "Pascal, Pa.")
```

This result may be assessed through details obtained from a literature review where it was found that glaciers typically have a shear stress of about $10^5 Pa$ at the base <sup>[17](https://www.tandfonline.com/doi/pdf/10.4296/cwrj3602823?needAccess=true)</sup> . 

### Creep Exponent
A value of 1 equals linear flow. As the value tends to infinity, it corresponds to perfect plastic. Values are generally between 1.9 and 4.5 <sup>[16](https://www.geo.umass.edu/courses/geo563/Geo563-6.IceFlow&profiles.pdf)</sup>.

```python
# Generate 100 data points from normal distribution between approx. 1.9 and 4.5
n = np.random.normal(3.2,0.52,100)

# Print max and min values to screen
print("The max and min values generated for creep exponent are %3.2f" %np.max(n),"and %3.2f" %np.min(n),".")
```

### Strain Rate

To calculate strain rate, ε, from Glen's Flow Law:

ε = B $\sigma ^ n$ 

Where:

ε = Strain Rate

B = Temperature dependent ice hardness parameter

$\sigma$ = Basal Shear Stress

n = Creep Exponent

```python
# Calculate strain rate from Glen's Flow Law
ε = B*(np.abs(σ)*(10**-6))**n
```

```python
# Plot the values of strain rate against stress to visualise
# Logarithmic scale used due to scale
plt.scatter(np.log10(σ),np.log10(ε));
```

```python
# Add strain rate to generated data set
data['Strain Rate'] = ε
```
